```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
# Nombro mi dataframe
df = pd.read_csv('/content/variables_dummies_score.csv')
df2 = pd.read_csv('/content/consolidado_final_entrega.csv')
```

```python
#Calcular matriz de confusión Primero Modelo
# Modelo sin pesos en las variables dummies
# Score de 0 a 7, suma de 7 variables dummies
y_true = df['default_flag']
y_pred = df['flag_pagador']
cm = confusion_matrix(y_true, y_pred, labels=[0, 1])

# Imprimir matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Buen pagador', 'Mal pagador'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusión')
plt.show()

# Calcular y mostrar el reporte de clasificación
report = classification_report(y_true, y_pred, labels=[0, 1], target_names=['Buen Pagador', 'Mal Pagador'])
print('Reporte de Clasificación:')
print(report)

# Calcular métricas
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label=0)
recall = recall_score(y_true, y_pred, pos_label=0)
f1 = f1_score(y_true, y_pred, pos_label=0)

# Imprimir métricas
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```
